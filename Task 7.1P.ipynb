```python
!pip install pandas numpy seaborn
!pip install seaborn plotly

import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("DHT22.csv", header=0, sep=",")
X = data.Temperature.values.reshape(-1, 1)

model = LinearRegression()
model.fit(X, data.Humidity)

x_range = np.linspace(X.min(), X.max(), 100)
y_range = model.predict(x_range.reshape(-1, 1))

fig = px.scatter(data, x='Temperature', y='Humidity', opacity=0.65)
fig.add_traces(go.Scatter(x=x_range, y=y_range, name='Regression Fit'))
fig.show()
```

```python
df1 = data[(data['Temperature'] > 34.8) & (data['Temperature'] < 37) & (data['Humidity'] > 23.5) & (data['Humidity'] < 23.8)]

fig = px.scatter(df1, x='Temperature', y='Humidity', opacity=0.65)
fig.add_traces(go.Scatter(x=x_range, y=y_range, name='Regression Fit'))
fig.show()
```
